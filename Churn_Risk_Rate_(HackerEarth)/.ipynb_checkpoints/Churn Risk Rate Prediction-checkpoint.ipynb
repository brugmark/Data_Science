#### Outliers
Now, we will look for potential outliers in the numeric data. These are identified by calculating the z-score. This represents how many standard deviations a value deviates from the mean. The farther away, the more likely the value is an outlier. In the plots below, a value with a z-score of 3 or more is colored red. Please keep in mind that red indicates a potential outlier which is not necessarily an actual outlier!

Looking at the scatterplots below, it becomes clear that the variables 'days_since_last_login', 'avg_time_spent', and 'churn_risk_score' contain "outliers" for sure. This can be concluded because some of the values not only deviate a lot from the mean, but are also negative which is not realistic looking at the nature of the variables (e.g. time spent cannot be negative).

```python
# Check potential outliers
dic_out = {}
for col in data_num.columns:
    data_nna = data_num[data_num[col].notna()]
    data_nna.reset_index(drop=True, inplace=True)
    out_data = np.array(data_nna[col])
    z_score = abs(stats.zscore(out_data))
    out_index = np.where(z_score > 3)
    out_count = len(out_index[0])
    dic_out[col] = {}
    dic_out[col]['values'] = data_num[col].dropna()
    dic_out[col]['z_score'] = z_score
    dic_out[col]['out_index'] = list(out_index[0])
    dic_out[col]['out_count '] = out_count
    # Plot
    fig = plt.figure(col + ' outliers')
    ax0 = fig.add_subplot(1, 1, 1)
    data_nna.plot.scatter('churn_risk_score', col, color='b', ax=ax0)
    data_nna.loc[dic_out[col]['out_index'], :].plot.scatter('churn_risk_score', col, color='r', ax=ax0)
```

### Reformat
Some of the data ('joining_date' and 'last_visit_time') need to be reformatted to numerical data to allow the algorithm to "read" these data. 

```python
data['joining_date'] = pd.to_datetime(data['joining_date'])
latest_date = max(data['joining_date'])
data['joining_date'] = latest_date - data['joining_date']
data['joining_date'] = data['joining_date'].dt.days
data['joining_date']
```

```python
data['last_visit_time'] = [t.split(':') for t in data['last_visit_time']]
lvt = list()
for ls in data['last_visit_time']:
    lvt.append([int(x) for x in ls])
data['last_visit_time'] = [np.dot(ls, [3600, 60, 1]) for ls in lvt]
data['last_visit_time']
```

In addition, there are binary and ordinal variables that can be transformed to numeric values.

```python
dic_map = {}
dic_map['gender'] = {'F': 0, 'M': 1, 'Unknown': float('NaN')}
dic_map['region_category'] = {'Village': 1, 'Town': 2, 'City': 3}
dic_map['membership_category'] = {'No Membership': 1, 
                                  'Basic Membership': 2, 
                                  'Silver Membership': 3, 
                                  'Gold Membership': 4, 
                                  'Platinum Membership': 5, 
                                  'Premium Membership': 6}
dic_map['joined_through_referral'] = {'No': 0, 'Yes': 1}
dic_map['used_special_discount'] = {'No': 0, 'Yes': 1}
dic_map['offer_application_preference'] = {'No': 0, 'Yes': 1}
dic_map['past_complaint'] = {'No': 0, 'Yes': 1}
dic_map['feedback'] = {'No reason specified': 0,  # negative
                       'Poor Customer Service': 0,
                       'Poor Product Quality': 0,
                       'Poor Website': 0,
                       'Too many ads': 0,
                       'Products always in Stock': 1,
                       'Quality Customer Care': 1,
                       'Reasonable Price': 1,
                       'User Friendly Website': 1
                      }  # positive

for var in dic_map:
    data[var] = data[var].map(dic_map[var])
```

And nominal variables that can be transformed to dummy variables.

```python
nom_vars = ['preferred_offer_types', 'medium_of_operation', 'internet_option', 'complaint_status']
data_nom = pd.get_dummies(data.loc[:, nom_vars])
data.drop(nom_vars, axis=1, inplace=True)
data = pd.concat([data, data_nom], axis=1)
data = data.reindex(sorted(data.columns, key=str.casefold), axis=1)
data.info()
```

### Clean

Outlier removal (see outlier plots in section Exploration)

```python
for column in ['days_since_last_login', 'avg_time_spent', 'avg_frequency_login_days', 'points_in_wallet']:
    data[column] = [float('NaN') if value < 0 else value for value in data[column]]
data = data.loc[data['churn_risk_score'] != -1]
```

Impute missing values ("NaN")

```python
data_dv = data['churn_risk_score']
data_cols = data.columns
imputer = IterativeImputer()
imputer.fit(data)
data = pd.DataFrame(imputer.transform(data))
data.columns = data_cols
data['churn_risk_score'] = data_dv
```

Round imputed float values that should be integers

```python
for column in ['joined_through_referral', 'region_category', 'avg_frequency_login_days', 'gender']:
    data[column] = [round(value) for value in data[column]]
```

# 4. Modeling

Now the actual modeling will start. The following algorithms will be applied:
- Kernel SVM
- Random Forest
- Neural Network
- Gradient Boosting Tree
- Extreme Gradient Boosting

For each algorithm the performance (f1_macro-score) will be determined using K-fold (5) cross validation. 

Furthermore, the model will be improved using:
- Confusion Matrix
- Learning Curves
- Feature Selection

```python
dic_scores = {}
dic_estimators = {}
score = 'f1_macro'  # https://www.kaggle.com/enforcer007/what-is-micro-averaged-f1-score
```

Kernel SVM

```python
# https://towardsdatascience.com/svm-and-kernel-svm-fed02bef1200
svm_par_grid = {
    "kernel": ['rbf'],  # https://data-flair.training/blogs/svm-kernel-functions/
}
svm = SVC(random_state=1)
svm_gs = GridSearchCV(estimator=svm, param_grid=svm_par_grid, n_jobs=-1)
svm_gs.fit(train_idv, train_dv)
print(svm_gs.best_params_)
svm = SVC(**svm_gs.best_params_, random_state=1)
dic_scores['Kernel SVM'] = cross_val_score(svm, train_idv, train_dv, cv=5, scoring=score).mean()
dic_estimators['Kernel SVM'] = svm
```

```python
dic_scores['Kernel SVM']
```

Random Forest

```python
rf_par_grid = {
    "max_depth": [50, 100],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [2, 5],
    "n_estimators": [50, 100],
}
rf = RandomForestClassifier(random_state=1)
rf_gs = GridSearchCV(estimator=rf, param_grid=rf_par_grid, n_jobs=-1)
rf_gs.fit(train_idv, train_dv)
print(rf_gs.best_params_)
rf = RandomForestClassifier(**rf_gs.best_params_, random_state=1)
dic_scores['Random Forest'] = cross_val_score(rf, train_idv, train_dv, cv=5, scoring=score).mean()
dic_estimators['Random Forest'] = rf
```

```python
dic_scores['Random Forest']
```

Gradient Boosting Tree

```python
gbt_par_grid = {
    "max_depth": [50, 100],
    "min_samples_split": [2, 5],
    "min_samples_leaf":  [2, 5],
    "n_estimators": [50, 100],
    "learning_rate": [0.01, 1],
    }
gbt = GradientBoostingClassifier(random_state=1)
gbt_gs = GridSearchCV(estimator=gbt, param_grid=gbt_par_grid, n_jobs=-1)
gbt_gs.fit(train_idv, train_dv)
print(gbt_gs.best_params_)
gbt = GradientBoostingClassifier(**gbt_gs.best_params_, random_state=1)
dic_scores['Gradient Boosting Tree'] = cross_val_score(gbt, train_idv, train_dv, cv=5, scoring=score).mean()
dic_estimators['Gradient Boosting Tree'] = gbt
```

```python
dic_scores['Gradient Boosting Tree']
```

Extreme Gradient Boosting

```python
xgbr_par_grid = {
    "max_depth": [50, 100],
    "n_estimators": [50, 100],
    "learning_rate": [0.01, 1],
    "reg_alpha": [0.01, 1],
    "reg_lambda": [0.01, 1],
}
xgbr = XGBClassifier(objective='multi:softmax', random_state=1)
xgbr_gs = GridSearchCV(estimator=xgbr, param_grid=xgbr_par_grid, n_jobs=-1)
xgbr_gs.fit(train_idv, train_dv)
print(xgbr_gs.best_params_)
xgbr = XGBClassifier(**xgbr_gs.best_params_, objective='binary:hinge', random_state=1)
dic_scores['Extreme Gradient Boosting'] = cross_val_score(xgbr, train_idv, train_dv, cv=5, scoring=score).mean()
dic_estimators['Extreme Gradient Boosting'] = xgbr
```

```python
dic_scores['Extreme Gradient Boosting']
```

### Learning Curves

Plot learning curves (Bias or Variance?)

```python
# Set-up
train_sizes = [2, 10, 50, 250, 1250, 6250, 28663]
train_sizes, train_scores, validation_scores = learning_curve(estimator=estimator, X=train_idv, y=train_dv,
                                                              train_sizes=train_sizes, cv=5,
                                                              scoring=score)
train_scores_mean = -train_scores.mean(axis=1)
validation_scores_mean = -validation_scores.mean(axis=1)
df_lc = pd.DataFrame()
df_lc['train_scores'] = pd.Series(train_scores_mean)
df_lc['validation_scores'] = pd.Series(validation_scores_mean)
df_lc.set_index(train_sizes, inplace=True)

# Plot
fig = plt.figure('Learning Curves')
ax0 = fig.add_subplot(1, 1, 1)
df_lc.plot(kind='line', ax=ax0)
plt.xlabel('Training set size')
plt.ylabel(score)
plt.title('Learning Curves')
plt.legend()
```
